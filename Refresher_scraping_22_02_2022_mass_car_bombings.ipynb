```python
import pandas as pd
df=pd.DataFrame(data)
df.columns=thead
df
```

```python
def wiki_table():
    thead = [i.text.strip() for i in table2.find_all('th')]
    data=[]
    for r in soup.select('table.wikitable.sortable.jquery-tablesorter')[1].tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```

```python
wiki_table()
```
